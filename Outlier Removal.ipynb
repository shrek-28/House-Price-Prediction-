### Library Imports


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading in the Dataset

```python
df = pd.read_csv('./DATA/Ames_Housing_Data.csv')
```

```python
df.head()
```

### Correlation Analysis

```python
plt.figure(figsize=(4,4))
sns.heatmap(df.corr(numeric_only=True))
plt.title("Correlation Analysis: Ames Housing Dataset");
```

```python
df.corr(numeric_only=True)['SalePrice'].sort_values()
```

### Viewing the Distribution: Sale Prices

```python
sns.histplot(df["SalePrice"], kde=True)
```

### Relationship between OverallQual (Rates the overall material and finish of the house) and Sale Price

```python
sns.scatterplot(x='Overall Qual',y='SalePrice',data=df)
```

#### Viewing the Outlier Rows, as inferred from the scatterplot

```python
df[(df['Overall Qual']>8) & (df['SalePrice']<200000)]
```

### Relationship between Gr Liv Area (Above grade (ground) living area square feet) and Sale Price

```python
sns.scatterplot(x='Gr Liv Area',y='SalePrice',data=df)
```

#### Viewing the Outlier Rows, as inferred from the scatterplot

```python
df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000)]
```

#### Identifying the indices of the Outlier Columns 

```python
df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000)].index
```

```python
ind_drop = df[(df['Gr Liv Area']>4000) & (df['SalePrice']<400000)].index
```

```python
df = df.drop(ind_drop,axis=0)
```

### Visualization post-dropping the Outliers

```python
sns.scatterplot(x='Gr Liv Area',y='SalePrice',data=df)
```

```python
sns.scatterplot(x='Overall Qual',y='SalePrice',data=df)
```

### Conversion to CSV post-outlier dropping

```python
df.to_csv("./DATA/Ames_outliers_removed.csv",index=False)
```
