#### Visualizing the Percentage of Missing Values in each Column of the DataFrame

````python
def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan

percent_nan = percent_missing(df)
````

````python
plt.figure(figsize=(15, 5))  # Wider horizontal layout
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

#### Visualizing this graph with a 1% threshold

````python
plt.figure(figsize=(15, 5))  # Wider horizontal layout
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.ylim(0,1)
plt.tight_layout()
plt.show()
````

### Dealing with the Basement Columns

````python
df[df['Total Bsmt SF'].isnull()]
````

````python
df[df['Bsmt Half Bath'].isnull()]
````

#### Working with numerical basement features 

````python
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
````

#### Working with String Basement features

````python
bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')
````

#### Reviewing the Dataset post-filling in columns 

````python
percent_nan = percent_missing(df)
````

````python
plt.figure(figsize=(15, 5))  # Wider horizontal layout
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

### Dropping Rows 

#### The ```Electrical``` and ```Garage Cars``` column, have very less null values, which account for only one or two rows. Hence they can be dropped. 

````python
df = df.dropna(axis=0,subset= ['Electrical','Garage Cars'])
````

#### Re-visualization post-dropping of rows 

````python
percent_nan = percent_missing(df)
````

````python
plt.figure(figsize=(15, 5))  
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

### Mass Vnr Feature

Based on the Description Text File, Mas Vnr Type and Mas Vnr Area being missing (NaN) is likely to mean the house simply just doesn't have a masonry veneer, in which case, we will fill in this data as we did before.

````python
df["Mas Vnr Type"] = df["Mas Vnr Type"].fillna("None")
df["Mas Vnr Area"] = df["Mas Vnr Area"].fillna(0)
````

#### Re-visualization post-filling on of Data

````python
percent_nan = percent_missing(df)

plt.figure(figsize=(15, 5))  
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

## Imputation of Missing Data

Neighborhood: Physical locations within Ames city limits
LotFrontage: Linear feet of street connected to property
It was operated under the assumption, that Neighbourhood is correlated to lot frontage. 

````python
df['Neighborhood'].unique()
````

````python
plt.figure(figsize=(8, 12))
sns.boxplot(
    x='Lot Frontage',
    y='Neighborhood',
    data=df,
    orient='h',
    palette='Set3'  
)
plt.tight_layout()
plt.show()
````

````python
df.groupby('Neighborhood')['Lot Frontage']
````

````python
df.groupby('Neighborhood')['Lot Frontage'].mean()
````

#### Column Transformation using Mean Values 

````python
df.head()['Lot Frontage']
````

````python
df[df['Lot Frontage'].isnull()]
````

````python
df.iloc[21:26]['Lot Frontage']
````

````python
df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean()))
````

````python
df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean())).iloc[21:26]
````

````python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean()))
````

#### Re-visualization of Data, post treatment of Lot Frontage Column

````python
percent_nan = percent_missing(df)

plt.figure(figsize=(15, 5))  
sns.barplot(x=percent_nan.index, y=percent_nan, color='pink')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
````

````python
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
````
